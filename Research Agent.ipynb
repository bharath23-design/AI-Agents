# import the packages

```python
from phi.agent import Agent , RunResponse
from phi.model.groq import Groq
from phi.tools.duckduckgo import DuckDuckGo
from phi.tools.newspaper4k import Newspaper4k 
import markdown
from IPython.display import display,HTML
import os
from dotenv import load_dotenv
load_dotenv()
```

# What is
- duckduckgo

- newspaper4k tools

# Build AI Research Agent

```python
Research_agent= Agent(
    model=Groq(id="gemma2-9b-it"),
    tools=[DuckDuckGo(), Newspaper4k()],
   description = "You are an expert financial analyst and journalist specializing in market trends, economic policies, and investment insights.",
    instructions = [
        "For a given financial or economic topic, search for the top 5 authoritative sources (news, research reports, and expert analysis).",
        "Extract key insights and summarize the main takeaways from each source.",
        "Analyze the information, compare different perspectives, and ensure accuracy in reporting.",
        "Write a well-structured, in-depth article with clear sections, factual data, and professional analysis.",
        "Ensure the article is engaging, informative, and suitable for publication in a leading financial journal or news outlet."
    ],
    markdown=True,
    show_tool_calls=True,
    add_datetime_to_instructions=True,
)
```

```python
Research_agent
```

```python
run: RunResponse = Research_agent.run("Write an Article about the IndianStock market")

text = run.content
```

```python
text
```

```python
md=markdown.Markdown()
markdown_content=md.convert(text)
markdown_content
```

```python
display(HTML(markdown_content))
```
